- Matching ihrer Tonart (Key)

## Beat Matching
Das Matching der Geschwindigkeit (Beat Matching) wird von gängiger Software automatisch unterstützt und hat eine geringe Fehlerquote. Das Feststellen des Beats ist durch die betonte, vordergründige Amplitude der Bassdrum/Kick leicht.

```python
Image(filename='./jupyter/bpmbestimmung.png')
# Die Basedrum (50 -200 Hz) ist das kräftigste Element bei Elektronischer Musik. Ihre Transiente kann leicht für die BPM Bestimmung herangezogen werden.
```

Ist die Geschwindigkeit/ BPM = Beats per Minute zweier Musikstücke bekannt, können diese im Takt, d.h. ohne Versatz nebeneinander gespielt werden, so dass ein gefälliger Übergang entsteht, ohne dass eine Pause erforderlich ist.

## Key Matching
Beim Key Matching wird zusätzlich zum Beat Matching versucht, die gleiche Tonart oder zumindest den gleichen Grundton zweier Musikstücke zu treffen. Dabei können noch glattere und unauffälligere Übergänge erzeugt werden. Mehr noch ist ein nicht störender Übergang zwischen zwei Musikstücken, deren Tonarten nicht zusammenpassen, kaum möglich, wodurch man in der Wahl der Musikstücke eingeschränkt wird.

```python
Image(filename='./jupyter/bpmandkey.png')
# Grün: BPM - Beats per Minute; Rot: Tonart - Key
```

Die Abspielgeschwindigkeit, wie auch die Tonhöhe eines Musikstücks können heutzutage unabhängig voneinander eingestellt werden, jedoch nicht beliebig weit. Mehr als drei Ganztonintervalle sind nicht ohne eine hörbare Verschlechterung der Klangdynamik möglich.

Außerdem arbeiten Analysetools nicht fehlerfrei, was es erforderlich macht trotzdem jede vorhergesagte Tonart manuell zu überprüfen, wenn man im Live Einsatz keine Überaschung erleben möchte.
