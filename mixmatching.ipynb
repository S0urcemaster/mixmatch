## Erweitern des Spektrums

Es gibt also 12 mögliche Grundtöne mit eher vernachlässigbaren Obertönen.

Ein DJ Set von 2 Stunden benötigt bei einer Spielzeit von höchstens 5 Minuten pro Track, mindestens 24 Tracks.

Wenn man sich nur auf einen Grundton beschränkt, kann die Auswahl eng werden bzw. man ist weniger frei, bestimmte Tracks zu spielen. Es wird also besonders nach Matches gesucht, die nicht auf demselben Grundton liegen, aber trotzdem harmonische Übergänge erlauben.

Es soll also eine Möglichkeit geschaffen werden, individuelle Matches verknüpfen zu können und aufgrund dessen, möglicherweise weitere Lösungen finden zu können.

Wenn ein Match außerhalb der gleichen Tonart (oder des Quintenzirkels) gefunden wird, muss es auch übertragbare Kriterien dafür geben. Davon auszuschließen ist natürlich, dass die zuvor bestimmte Tonart falsch ist.
